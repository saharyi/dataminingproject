```python
''''From the statistical result, it is observed that  glucose , blood pressure, Skin thickness,insulin and BMI
have a min value of 0.Medical knowledge explains that such attributes (medical result) cannot be 0;
theref that the dataset contains a missing value that if not handled can impair the quality of our model result
and accuracy. Various methods have been suggested for handling missing values in datasets. In
our case we replaced mis ing values with the mean such attribute.'''

data['Glucose'] = data['Glucose'].replace(0, 120.89)
data['BloodPressure'] = data['BloodPressure'].replace(0, 69.10)
data['SkinThickness'] = data['SkinThickness'].replace(0, 20.53)
data['Insulin'] = data['Insulin'].replace(0, 79.79)
data['BMI'] = data['BMI'].replace(0,31.99)

print(data.min())

display(data)
```

```python
#Using PCA to dimensionality reduction

   

from sklearn import preprocessing

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from scipy.stats.stats import pearsonr

X2= data.drop('Outcome', 1)
y2= data['Outcome']

scale2= StandardScaler()

scaled_data2 = scale2.fit_transform(X2)

from numpy.linalg import eigh
cov_matrix = np.cov(scaled_data2, rowvar=False)
#
# Determine eigenvalues(how much each component  have varianse )
#and eigenvectors(vectors that used for creat component)
#
egnvalues, egnvectors = eigh(cov_matrix)
#
# Determine explained variance
#
total_egnvalues = sum(egnvalues)
var_exp = [(i/total_egnvalues) for i in sorted(egnvalues, reverse=True)]
#
# Plot the explained variance against cumulative explained variance
#
import matplotlib.pyplot as plt
cum_sum_exp = np.cumsum(var_exp)

print(cum_sum_exp)
plt.bar(range(0,len(var_exp)), var_exp, alpha=0.5, align='center', label='Individual explained variance')
plt.step(range(0,len(cum_sum_exp)), cum_sum_exp, where='mid',label='Cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component index')
plt.legend(loc='best')
plt.tight_layout()
plt.show()

from sklearn.decomposition import PCA
pca = PCA(n_components=4)
principalComponents = pca.fit_transform(scaled_data2)

_d2= pd.DataFrame(principalComponents)
_d2.columns = ['x', 'y','z','t']
_d2['Outcome']=y2
_d2.to_csv("2d-PCA_new.csv")

f2 = pd.read_csv("2d-PCA_new.csv", index_col=0)

print(f2)




''''covariance_matrix2 = np.cov(scaled_data2 .T)
eigen_values2, eigen_vectors2 = np.linalg.eig(covariance_matrix2)

##print("Eigenvector: \n",eigen_vectors,"\n")

##print("Eigenvalues: \n", eigen_values, "\n")

# Using two first components (because those explain more than 95%)
projection_matrix2 = (eigen_vectors2.T[:][:2]).T


X_pca2 = X2.dot(projection_matrix2)

_d2= pd.DataFrame(X_pca2)
_d2.columns = ['x', 'y']
_d2['Outcome']=y2
_d2.to_csv("2d-PCA_new.csv")

f2 = pd.read_csv("2d-PCA_new.csv", index_col=0)
print(f2)'''






```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
# composing data set
ddf = pd.read_csv("2d-PCA_new.csv", index_col=0)
Xd = ddf.drop('Outcome', 1)
yd = ddf['Outcome']

xxd=np.array(Xd)

km = KMeans(n_clusters = 2)

clusters=km.fit_predict(xxd)

# plotting data set
##plt.scatter(*zip(*xxd),c=clusters,marker = 'x')

# obtaining the centers of the clusters
centroids = km.cluster_centers_

# points array will be used to reach the index easy
points = np.empty((0,len(xxd[0])), float)
# distances will be used to calculate outliers

distances = np.empty((0,len(xxd[0])), float)
# getting points and distances

for i, center_elem in enumerate(centroids):
    # cdist is used to calculate the distance between center and other points
    distances = np.append(distances, cdist([center_elem],xxd[clusters == i], 'euclidean')) 
    points = np.append(points, xxd[clusters == i], axis=0)
    
percentile = 80
# getting outliers whose distances are greater than some percentile
outliers = points[np.where(distances > np.percentile(distances, percentile))]

print(len(outliers))

```

```python

df_22= pd.read_csv("2d-PCA_new.csv")

num=list()

for indexx in outliers:
    #print('index',indexx[0])
    num.append((df_22[(df_22.x == indexx[0]) & (df_22.y == indexx[1]) & (df_22.t == indexx[3])].index)[0]) 
    
df_22=df_22.drop(num,0)
df_22
```

```python
cleaned_data1=df_22['x']
cleaned_data2=df_22['y']
cleaned_data4=df_22['z']
cleaned_data5=df_22['t']

cleaned_data3=df_22['Outcome']

cleaned1=pd.DataFrame(cleaned_data1)
cleaned2=pd.DataFrame(cleaned_data2)
cleaned4=pd.DataFrame(cleaned_data4)
cleaned5=pd.DataFrame(cleaned_data5)

cleaned3=pd.DataFrame(cleaned_data3)

numbers = cleaned2["y"]
numbers4 = cleaned4["z"]
numbers5 = cleaned5["t"]

numbers2=cleaned3["Outcome"]

cleaned1 = cleaned1.join(numbers)
cleaned1 = cleaned1.join(numbers4)
cleaned1 = cleaned1.join(numbers5)

cleaned1=cleaned1.join(numbers2)

cleaned1 = cleaned1.reset_index()
cleaned1 = cleaned1.drop('index', 1)
print(cleaned1)
```

```python
#---------------------------Modeling with spiliting data to train and test ----------------------
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn import model_selection
from sklearn import metrics
from numpy import mean
from numpy import std

X1 = cleaned1.drop('Outcome',1)
y1 = cleaned1['Outcome']


X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size = 0.2,random_state=11)

classifier = LogisticRegression()

classifier.fit(X_train,  y_train)


predictions1 = classifier.predict(X_test)

scores = cross_val_score(classifier, X_test, y_test, scoring='accuracy', cv=10)


print('Accuracy of logistic regression classifier on test set: %.3f ' % (mean(scores)*100))



from sklearn.metrics import accuracy_score

print("Accuracy without kfold2 : %.2f"%(classifier.score(X_test, y_test)*100))
```

```python
 from sklearn.metrics import confusion_matrix



unique_label = np.unique([ y_test, predictions1])

cmtx = pd.DataFrame(
    confusion_matrix(y_test, predictions1, labels=unique_label), 
    index=['true:{:}'.format(x) for x in unique_label], 
    columns=['pred:{:}'.format(x) for x in unique_label]
)

print(cmtx)
```

```python
from sklearn.metrics import classification_report
target_names = ['class 0', 'class 1']
print(classification_report(y_test, predictions1,target_names=target_names))




```

```python
from sklearn.metrics import matthews_corrcoef
print(matthews_corrcoef(y_test, predictions1))
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test, predictions1))
```

```python
#ROC is a probability curve and AUC represents the degree or measure of separability.
#Higher the AUC, the better the model is at predicting 0 classes as 0 and 1 classes as 1

from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

false_positive_rate1, true_positive_rate1, threshold1 = roc_curve(y_test, predictions1)

plt.subplots(1, figsize=(10,10))
plt.title('Receiver Operating Characteristic - LogisticRegression')
plt.plot(false_positive_rate1, true_positive_rate1)
plt.plot([0, 1], ls="--")

plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
from sklearn.metrics import cohen_kappa_score

#Kappa = (observed accuracy - expected accuracy)/(1 - expected accuracy)
#observed accuracy=number of instance that grandtruth and classifier agreement about them =(357+105/614=0.75)

#expected accuracy=number of instance that each of them say it belong to a class for example grandtruth say
#357+53=410 belong to class 0 and classifier say 357+99=456 belong to class 0 so expected agreement for 
#class 0 is (410/614)*(456/614)=0.48  and for class 1 is (204/614)*(158/614)=0.08 
#so the expected accuracy is=(0.48+0.08)=0.56 and kappa is:
#(0.75-0.56)/(1-0.56)=0.40 
#numbers changed because accuracy and model and confusion matrix changhed



kappa = cohen_kappa_score(y_test, predictions1)
print('Cohens kappa: %f' % kappa)
```
